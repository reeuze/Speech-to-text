```python
import speech_recognition as sr
from difflib import SequenceMatcher
import os
```

```python
def audio_to_text(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data)
            return text
        except sr.UnknownValueError:
            return ""
        except sr.RequestError:
            return ""
```

```python
Target_words = ["City"]
```

```python
def similarity_ratio(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio()
```

```python
def Compare_selected_words(audio_files, target_word):
    comparison_results = []

    for audio_file in audio_files:
        recognized_words = audio_to_text(audio_file)

        similarities = [round(similarity_ratio(target_word, recognized_words), 5)]
        comparison_results.append(similarities)
    
    return comparison_results
```

```python
def get_all_audio_files(root_folder):
    audio_files = []
    for i in range(1, 26):
        folder_name = f"Data-{i}"
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            for subdir, _, files in os.walk(folder_path):
                for file in files:
                    file_path = os.path.join(subdir, file)
                    audio_files.append(file_path)
    return audio_files

audio_files = []
```
